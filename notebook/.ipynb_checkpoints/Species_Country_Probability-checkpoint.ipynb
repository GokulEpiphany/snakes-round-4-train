```python
#Given a country, what is the probability of finding a particular species
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('species_per_country_count.csv',low_memory=False)
```

```python
df.head()
```

```python
df_needed = df.iloc[:,1:]
```

```python
df_needed
```

```python
prob_from_df = np.array(df_needed.values)
```

```python
probability_per_country=prob_from_df/prob_from_df.sum(axis=1, keepdims=True)
```

```python
df.iloc[:,1:]=probability_per_country
```

```python
df.head()
```

```python
df.to_csv('probability_of_species_per_country.csv',index=True)
```
