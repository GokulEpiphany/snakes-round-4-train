```python
names = train_df['scientific_name'] +'-'+ train_df['country']
```

```python
country_and_species = []
for name in names:
    country_and_species.append(name.lower().replace(' ','-'))
```

```python
country_wise_count=dict(zip(*np.unique(country_and_species,return_counts=True)))
```

```python
country_wise_count
```

```python
species_names = list(occurences.keys())
```

```python
country_count = train_df['country']
```

```python
country_inc = []
for country in country_count:
    country_inc.append(country.lower().replace(' ','-'))
```

```python
country_occurencees = dict(zip(*np.unique(country_inc,return_counts=True)))
```

```python
country_occurencees
```

```python
country_names=list(country_occurencees.keys())
```

```python
len(country_names)
```

```python
results_df = pd.DataFrame(columns=species_names)
```

```python
for idx,country in enumerate(country_occurencees):
    print("Saving info for country {}".format(country))
    for species in species_names:
        name=(species+'-'+country)
        try:
            results_df.loc[idx,species]=country_wise_count[name]
        except:
            results_df.loc[idx,species]=0
            pass
        
```

```python
results_df.insert(0, column='Species/Country', value=country_names)
```
